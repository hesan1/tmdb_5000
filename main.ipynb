Step 2: Data Cleaning and Preprocessing


```python
credits_df.rename(columns={'movie_id': 'id'}, inplace=True)
```

```python
merged_df = pd.merge(movies_df, credits_df, on='id')
```

```python
# Making a copy of movies dataset, using 'movies1' for EDA and 'movies' for recommender system
movies_eda= merged_df.copy()
```

```python
# Dropping unncessary columns
movies_eda = movies_eda.drop(columns=[
    'homepage', 'tagline', 'id', 'overview', 'status', 'original_title',])
```

```python
# Duartion of the data
movies_eda['release_date'] = pd.to_datetime(movies_eda['release_date'])
print(movies_eda['release_date'].max()-movies_eda['release_date'].min())
```

```python
movies_eda.info()
```

```python
import ast
```

```python
# Tidying up genre, production_companies and production_countries column
def func(obj):
    List = []
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List
```

```python
movies_eda['genres'] = movies_eda['genres'].apply(func)
movies_eda['production_companies'] = movies_eda['production_companies'].apply(func)
movies_eda['production_countries'] = movies_eda['production_countries'].apply(func)
```

Create Profitability Indicator

```python
movies_eda['profit'] = movies_eda['revenue'] - movies_eda['budget']
movies_eda['is_profitable'] = movies_eda['profit'] > 0
```

```python
movies_eda['genres'] = movies_eda['genres'].apply(lambda x: [i.lower() for i in x])
```

```python
# Removing rows with any null values
movies_eda.dropna(inplace=True)
```
